## Task 2 

### Task 2.1: importing covid19 dataset
importing "Covid19_Confirmed_dataset.csv" from "./Dataset" folder. 


```python
covid_19_con_cases=pd.read_csv('Datasets/covid19_Confirmed_dataset.csv')
```

#### Let's check the shape of the dataframe

```python
covid_19_con_cases    
```

```python
covid_19_con_cases.drop(['Lat','Long'],axis=1,inplace=True)
```

```python
covid_19_con_cases
```

### Task 2.3: Aggregating the rows by the country

```python
covid_19_agg=covid_19_con_cases.groupby('Country/Region').sum()
```

```python
covid_19_agg
```

### Task3: Calculating a good measure 
we need to find a good measure reperestend as a number, describing the spread of the virus in a country. 

```python
covid_19_agg.loc['China'].plot()
```

### task 3.1: caculating the first derivative of the curve

```python
covid_19_agg.loc['China'].diff().plot()
```

### task 3.2: find maxmimum infection rate for China

```python
covid_19_agg.loc['China'].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
max_infection=[]
for c in list(covid_19_agg.index):
    max_infection.append(covid_19_agg.loc[c].diff().max())
covid_19_agg['max_infection']=max_infection
```

```python
covid_19_agg
```

### Task 3.4: create a new dataframe with only needed column 

```python
covid_19_agg1=pd.DataFrame(covid_19_agg['max_infection'])
```

```python
covid_19_agg1
```
